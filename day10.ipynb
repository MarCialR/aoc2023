```python
start_test = (1,1)
start_test2 = (2,0)
start_puzzle = (111,14)

test2 = """
..F7.
.FJ|.
FJ.L7
|F--J
LJ...""".splitlines()[1:]

PIPES = {
"|":[(-1,0),(1,0)], #is a vertical pipe connecting north and south.
"-":[(0,-1),(0,1)], #is a horizontal pipe connecting east and west.
"L":[(-1,0),(0,1)], #is a 90-degree bend connecting north and east.
"J":[(-1,0),(0,-1)], #is a 90-degree bend connecting north and west.
"7":[(0,-1),(1,0)], #is a 90-degree bend connecting south and west.
"F":[(0,1),(1,0)]} #is a 90-degree bend connecting south and east.

def find_next(p, data, n):
    char = data[n[0]][n[1]] 
    pipe_nexts = PIPES[char]
    for possible in pipe_nexts:
        nn = (n[0]+possible[0], n[1]+possible[1])
        if not nn == p:
            return nn

def solution_1(data, start):
    previous = start
    next = (start[0],start[1]+1)
    steps = 0
    while not next == start:
        next_next = find_next(previous, data, next)
        previous = next
        next = next_next
        steps += 1
    solution = int((steps+1)/2)
    return solution

assert solution_1 (puzzle.test, start_test)  == 4
assert solution_1 (test2, start_test2) == 8
assert solution_1 (puzzle.data, start_puzzle) == 6812

print( f"Solution 1:  it takes {solution_1 (puzzle.data, start_puzzle)} steps along the loop to get from the starting position to the point farthest from the starting position") # 13623 0.1s
```

```python
test3="""
.F----7F7F7F7F-7....
.|F--7||||||||FJ....
.||.FJ||||||||L7....
FJL7L7LJLJ||LJ.L-7..
L--J.L7...LJF7F-7L7.
....F-J..F7FJ|L7L7L7
....L7.F7||L7|.L7L7|
.....|FJLJ|FJ|F7|.LJ
....FJL-7.||.||||...
....L---J.LJ.LJLJ...""".splitlines()[1:]

start_test3 = (4,12)

test4="""
FF7F7F7F7F7F7F7F---7
L|LJ||||||||||||F--J
FL-7LJLJ||||||LJL-77
F--JF--7||LJLJ7F7FJ-
L---JF-JLJ.||-FJLJJ7
|F|F-JF---7F7-L7L|7|
|FFJF7L7F-JF7|JL---7
7-L-JL7||F7|L7F-7F7|
L.L7LFJ|||||FJL7||LJ
L7JLJL-JLJLJL--JLJ.L""".splitlines()[1:]

start_test4 = (0,3)


def build_path(data, start):
    path = set()
    previous = start
    next = (start[0],start[1]+1)
    while not next == start:
        next_next, turn = find_next2(previous, data, next)
        path.add(next_next)
        previous = next
        next = next_next
    next_next, turn = find_next2(previous, data, next)
    path.add(next_next)
    return path


def find_next2(p, data, n):
    # previous, data, next needed because we want to know return whether
    # it is going down
    char = data[n[0]][n[1]] 
    pipe_nexts = PIPES[char]
    for index, possible in enumerate(pipe_nexts):
        nn = (n[0]+possible[0], n[1]+possible[1])
        if not nn == p:
            if index == 1:
                return nn, pipe_nexts[1]
            return nn, -1*pipe_nexts[1]

def is_enclosed (data, path, coords):
    #check to the left
    left = 0
    first = None
    for col in list(range(0,coords[1]))[::-1]:
        next_to_check = (coords[0],col)
        if next_to_check in path:
            char = data[next_to_check[0]][next_to_check[1]]
            
            if char in "|J7":
                left +=1
                if char != '|' and not first:
                    first = char
            if char in "LF":
                if first == "J":
                    if char == 'L': left -= 1
                    else: pass

                elif first == "7":
                    if char == 'F': left -= 1
                    else: pass
                else:
                    pd(coords, next_to_check, char, first)
                    raise Exception ("something strange 1")
                first=None
    #check to the right
    right = 0
    first = None
    for col in list(range(coords[1]+1, len(data[0]))):
        next_to_check = (coords[0],col)
        if next_to_check in path:
            char = data[next_to_check[0]][next_to_check[1]]
            
            if char in "|LF":
                right +=1
                if char != '|' and not first:
                    first = char
            if char in "J7":
                if first == "L":
                    if char == 'J': right -= 1
                    else: pass

                elif first == "F":
                    if char == '7': right -= 1
                    else: pass
                else:
                    pd(coords, next_to_check, char, first  )
                    raise Exception ("something strange 1")
                first=None

    result = left%2 == 1 or right%2 == 1  # This is the key
    return result

def print_file(data, path, tiles_enclosed):
    map_dict = {'F': '┌', '7': '┐', 'L': '└', 'J': '┘'} 
    sol = []
    for row in range(len(data)):
        new_row = []
        for col in range(len(data[0])):
            char = data[row][col]
            if ((row,col)) in path:
                new_row.append(char if char not in map_dict else map_dict[char])
            elif ((row,col)) in tiles_enclosed:
                new_row.append('█')
            elif char == '.':
                new_row.append('•')
            else:
                new_row.append(" ")
        sol.append(new_row)
    with open("visual solution", "w") as f:
        f.write("\n".join([ "".join(line) for line in sol ]))
        f.write("\nTILES ENCLOSED\n" + str(tiles_enclosed))
    

def solution_2(data, start):
    path = build_path(data, start)

    tiles_enclosed = []
    print_file(data, path, tiles_enclosed)
    for row in range(len(data)):
        for col in range(len(data[0])):
            if not (row,col) in path:
                if is_enclosed(data, path,(row,col)):
                    tiles_enclosed.append((row,col))
    print_file(data, path, tiles_enclosed)
    return len(tiles_enclosed)

assert solution_2 (puzzle.test, start_test ) == 1
assert solution_2 (test2, start_test2) == 1
assert solution_2 (test3, start_test3) == 8
assert solution_2 (test4, start_test4) == 10
assert solution_2 (puzzle.data, start_puzzle) == 527

print( f"Solution 2:  There are {solution_2 (puzzle.data, start_puzzle)} tiles enclosed by the loop") # 0.25s
```
