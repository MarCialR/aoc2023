```python
def move(matrix, dir):
    repeat = True
    while repeat:
        repeat = False

        new_matrix = []
        current = matrix[0]
        width = len(current)
        for i in range(len(matrix)-1):
            next = matrix[i+1]
            new_array = ''
            new_current = ''
            for j in range(width):
                c = current[j]
                n = next[j]
                if c == '.' and n == 'O':
                    c = 'O'
                    n = '.'
                    repeat = True
                
                new_array += c
                new_current += n
            current = new_current
            new_matrix.append(new_array)
        new_matrix.append(current)
        if repeat:
            matrix = new_matrix

    return new_matrix
    

"""transposed = ["".join(x) for x in list(zip(*pattern))]
"""

def calculate_load(matrix, dir):
    sum = 0
    for i,r in enumerate(matrix):
        sum +=  r.count('O') * (len(matrix)-i)
    return sum

def solution_1(raw_data):
    #print ("\n".join(raw_data))
    moved = move (raw_data, 'up')
    #print("\n\n")
    #print ("\n".join(moved))

    
    return calculate_load(moved, 'up')


assert solution_1 (puzzle.test) == 136

print( f"Solution 1:  {solution_1 (puzzle.data)} is the sum of the load caused by all of the rounded rocks")

assert solution_1 (puzzle.data) == 103333
```

```python

import timeit


# run same code 5 times to get measurable data
n = 5

# calculate total execution time
result = timeit.timeit(stmt='solution_1 (puzzle.data)', globals=globals(), number=n)

# calculate the execution time
# get the average execution time
print(f"Execution time is {result / n} seconds")
```
