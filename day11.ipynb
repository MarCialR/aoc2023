```python

def find_double_spaces(data):
    rows = []
    for i, row in enumerate(data):
        if '#' in row :
            continue
        else:
            rows.append(i)
    cols = []
    for col in range(len(data[0])):
        is_there = False
        for row in range(len(data)):
            if data[row][col] == '#':
                is_there = True
        if not is_there:
            cols.append(col)
    return (rows, cols)

# print(find_double_spaces(puzzle.test))

def get_galaxies(data, exp_rows, exp_cols):
    galaxies = []
    actual_row =0
    for r,row in enumerate(data):
        if r in exp_rows:
            actual_row +=1
            continue
        
        actual_col = 0
        for c, elem in enumerate(data[r]):
            if c in exp_cols:
                actual_col +=1
                continue
            if data[r][c] == '#':
                galaxies.append ([ r+actual_row , c+actual_col])
    return galaxies

def distance(pair):
    return abs(pair[0][0]-pair[1][0]) + abs(pair[0][1]-pair[1][1])


def solution_1(space):

    double_spaces = find_double_spaces(space)
    galaxies_expanded = get_galaxies(space, *double_spaces)
    #print(galaxies_expanded)

    # All possible pairs in List
    # Using list comprehension + enumerate()
    galaxy_pairs = [(a, b) for idx, a in enumerate(galaxies_expanded) for b in galaxies_expanded[idx + 1:]]
    
    solution = sum(map(distance, galaxy_pairs))
    return solution


assert solution_1 (puzzle.test) ==374

print( f"Solution 1:  {solution_1 (puzzle.data)} is the sum of the shortest path between every pair of galaxies")
```

```python
def get_galaxies_expanded(data, exp_rows, exp_cols, expansion_factor):
    galaxies = []
    actual_row =0
    for r,row in enumerate(data):
        if r in exp_rows:
            actual_row += expansion_factor-1
            continue
        
        actual_col = 0
        for c, elem in enumerate(data[r]):
            if c in exp_cols:
                actual_col += expansion_factor-1
                continue
            if data[r][c] == '#':
                galaxies.append ([ r+actual_row , c+actual_col])
    return galaxies


def solution_2(space, expansion_factor):

    double_spaces = find_double_spaces(space)
    galaxies_expanded = get_galaxies_expanded(space, *double_spaces, expansion_factor)
    #print(galaxies_expanded)

    # All possible pairs in List
    # Using list comprehension + enumerate()
    galaxy_pairs = [(a, b) for idx, a in enumerate(galaxies_expanded) for b in galaxies_expanded[idx + 1:]]
    
    solution = sum(map(distance, galaxy_pairs))
    return solution


assert solution_2 (puzzle.test, 10) == 1030
assert solution_2 (puzzle.test, 100) == 8410

print( f"Solution 2:  {solution_2 (puzzle.data, 1000000)} is the sum of the shortest path between every pair of galaxies applying a factor of 1M")
```
