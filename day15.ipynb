```python
def HASH(string):
    current = 0
    for l in string:
        current += ord(l)
        current *= 17
        current = current % 256
    return current


def solution_1(raw_data):
    sum = 0
    for step in raw_data[0].split(','):
        sum += HASH(step)

    
    return sum


assert solution_1 (puzzle.test) == 1320

print( f"Solution 1:  {solution_1 (puzzle.data)} is the sum of the results")

assert solution_1 (puzzle.data) == 508498
```

```python
class Lens:
    def __init__(self, string, lens):
        self.string = string
        self.lens = lens
        self.next = None
    
    def __str__(self):
        next = ""
        if self.next:
            next = " -> " + repr(self.next)

        return self.string + "-" + self.lens + next
    
    def __repr__(self):
        next = ""
        #print (f"{self.string}_{self.next}")
        if self.next:
            next = " -> " + repr(self.next)

        return self.string + "-" + self.lens + next

    def inserAtEnd(self, string, lens):
        new_node = Lens(string, lens)

        current_node = self
        while(current_node.next):
            current_node = current_node.next
    
        current_node.next = new_node

        
    def remove_node(self, string):
        current_node = self
    
        # Check if the head node contains the specified string
        if current_node.string == string:
            return current_node.next
    
        while current_node.next and current_node.next.string != string:
            current_node = current_node.next

        if not current_node.next: #we are at the last one
            return self
        else: # We need to remove a middle or last Lens
            if current_node.next.next: #it is a middle one
                current_node.next = current_node.next.next
            else:
                current_node.next == None

        return self

    def replace(self, string, lens):
        e = self
        while e:
            if e.string == string:
                e.lens = lens
                return
            e = e.next
    



def HASH2(string):
    op = None
    lens = -1
    if '=' in string:
        string, lens = string.split('=')
        op = 'add'
        
    if '-' in string:
        string = string.split('-')[0]
        op = 'subtract'

    box = 0
    for l in string:
        box += ord(l)
        box *= 17
        box = box % 256
    return box, string, op, lens

def lensChain_value (lens):
    i = 1
    sol = i * int(lens.lens)
    #print (f"+ {i} * {int(lens.lens)}")
    while lens.next:
        lens = lens.next
        i += 1
        sol += i* int(lens.lens)
        #print (f"+ {i} * {int(lens.lens)}")
        
        
    return sol

def execute2 (step, solution):
    box, string, op, lens = HASH2(step)

    
    #print(f"\n{box}, {string}, {op}, {lens}")

    if op == 'subtract':
        if string in solution[box][0]:
            #print(f"{box}, {string}, {op}, {lens}")
            #print (f"1->{solution[box][0]}")

            solution[box][0].remove(string) #remove from set
            #print (f"2->{solution[box][0]}")
            solution[box][1] = solution[box][1].remove_node(string) #remove from linkedlist
        else:
            #print("Not substracting ANYTHING! ")
            pass
    else: #must be add/replace
        if string in solution[box][0]:
            #print("its replace")
            solution[box][1].replace(string,lens) #Replace!!
        else: #add first or at the end
            if not solution[box][0]:
                #print("its add first")
                solution[box][0] = set()
                solution[box][0].add(string)
                solution[box][1] = Lens(string, lens)
            else:
                #print("its ADDING at END!")
                solution[box][0].add(string)
                solution[box][1].inserAtEnd(string,lens)



def solution_2(raw_data):
    solution = [[set(),None] for i in range(256)]
    for step in raw_data[0].split(','):
        #print (f" \n executing {step}")
        execute2 (step, solution)
        #print(solution[:4])
    
    
    
    suma = 0
    for i, box in enumerate(solution):
        if box[0]:
            cuanto =  lensChain_value(box[1])
            suma += (i+1) * cuanto
            #print (f"en {box[0]} sumamos {cuanto}") 


    return suma



assert solution_2 (puzzle.test) == 145

print( f"Solution 2:  {solution_2 (puzzle.data)} is the sum of the results")

#assert solution_2 (puzzle.test) == 1320
```
