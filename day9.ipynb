```python

def parse_histories(raw_report):
    report = []
    for raw_history in raw_report:
        report.append(list(map(int,raw_history.split())))
    return report


def analyze_history(history):
    differences = []
    for i in range (len(history)-1):
        differences.append(history[i+1]-history[i])
    #print(differences)
    return differences


def recursive_analysis(history):
    if not any(history):  # I messed this one up here!
        return 0
    else:
        return history[-1] + recursive_analysis(analyze_history(history) )
    
def solution_1(raw_report):

    report = parse_histories(raw_report)
    
    #print(report)
    return sum(map(recursive_analysis, report))


assert solution_1 (puzzle.test) == 114

print( f"Solution 1:  {solution_1 (puzzle.data)} is the sum of the extrapolated values in the OASIS report")
```
