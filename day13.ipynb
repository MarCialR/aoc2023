```python

def parse_patterns(raw_data):

    patterns = []

    pattern = []
    for line in raw_data:
        if not line.strip(): # empty line
            patterns.append(pattern)
            pattern = []
            continue
        pattern.append(line)
    patterns.append(pattern)

    return patterns
    
def find_base_mirror(pattern):
    for i in range(len(pattern)-1):
        #print (f'analizing {i} - {i+1} - {pattern [i]} == {pattern [i+1]}')
        if pattern [i] == pattern [i+1]:
            #print (f'found  reflex in rows {i} - {i+1} - {pattern [i]} == {pattern [i+1]}')
            yield i, i+1
    
def are_reflection(pattern, l, r):

    left = pattern[:l+1]
    left.reverse()
    right = pattern[r:]
    for i in range (min(len(left), len(right))):
        if left[i] != right[i]:
            return False

    
    return True

    
def solution_1(raw_data):

    patterns = parse_patterns(raw_data)
    assert len(parse_patterns(puzzle.test)) ==2

    total = 0

    for i, pattern in enumerate(patterns):
        #print (f"\n\n ==============\nchecing Pattern {i}\n ==============")
        #print("\n".join(pattern))
        
        
        for l,r in find_base_mirror(pattern):
            #print (f'\nfound  reflex in rows {l} - {r}')
            if are_reflection(pattern, l, r):
                #print ("... and it is a good one!")
                #print (f"add {r} to horizontals!!!")
                total += r*100
        
        transposed = ["".join(x) for x in list(zip(*pattern))]
        
        #print("\nVERTICALS (HZ TRANSPOSED)\n ------------------")
        for l,r in find_base_mirror(transposed):
            #print (f'\nfound  reflex in rows {l} - {r}')
            if are_reflection(["".join(x) for x in list(zip(*pattern))], l, r):
                #print ("... and it is a good one!")
                #print (f"add {r} to verticals!!!")    
                total += r
    
    return total


assert solution_1 (puzzle.test) == 405

print( f"Solution 1:  {solution_1 (puzzle.data)} is the number I get after summarizing all of my notes")
```
